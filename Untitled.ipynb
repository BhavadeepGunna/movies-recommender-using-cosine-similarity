```python
df=movies[['genres','keywords','overview','production_companies','cast','crew','movie_id','title']]
```

```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
        
    return L
    

        
    
```

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head()
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df['production_companies']=df['production_companies'].apply(convert)
```

```python
df['cast']=df['cast'].apply(convert)
```

```python
df.head()
```

```python
df['cast']=df['cast'].apply(lambda x:x[0:3])
```

```python
df
```

```python
def get_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
df['crew']=df['crew'].apply(get_director)
```

```python
def change(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df.info()
```

```python
df['cast'][0]
```

```python
df['cast']=df['cast'].apply(change)
df['genres']=df['genres'].apply(change)
df['crew']=df['crew'].apply(change)
df['keywords']=df['keywords'].apply(change)
df['production_companies']=df['production_companies'].apply(change)

```

```python
df.head()
```

```python
def split_text(x):
    if isinstance(x, str):
        return x.split()
    else:
        return []
```

```python
df['overview']=df['overview'].apply(split_text)
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']+df['production_companies']
```

```python
df.head(1)
```

```python
new=df.drop(columns=['genres','keywords','overview','production_companies','cast','crew',])
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie']
```

```python
distances = sorted(list(enumerate(similarity[new[new['title'] == 'The Lego Movie'].index[0]])),reverse=True,key = lambda x: x[1])
```

```python
distances
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('The Lego Movie')
```
